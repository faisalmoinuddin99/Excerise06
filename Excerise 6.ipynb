```python
# snake Water Gun

# 10 baar challana hai using while loop

# snake water ya gun mai sai ek chunna hai variable mai store kr na hai

# ab computer sai random choose kr wana hai snake water ya gun. if computer ne snake liye or aap ne gun then YOU win

# scoring kr na hai

# jiska score 10 match mai subse zada hoga wo jeete ga

# or score bhi display kr wane hai
```

```python
import random
i = 1
element = ['Snake','Water','Gun']
player_score = 0
player_high_score = 0
computer_score = 0
computer_high_score = 0
while(i <= 5):
    computer_choice = random.choice(element)
    
    print("Please select the following elements:")
    print("1. Snake 2. Water 3. Gun")
    user_choice = input()
########### SNAKE ####################################    
    if (user_choice == '1') and (computer_choice == 'Water'):
        print("You win")
        player_score += 5
        if player_score > player_high_score:
            player_high_score = player_score
        
    
    if user_choice == '1' and computer_choice == 'Gun' :
        print("You Lost")
        computer_score += 5
        if computer_score > computer_high_score:
            computer_high_score = computer_score
    
    if user_choice == '1' and computer_choice == 'Snake':
        print("Tie")
        player_score += 1
        computer_score += 1
        if player_score > player_high_score or computer_score > computer_high_score :
            player_high_score = player_score
            computer_high_score = computer_score

########### WATER #################################### 
    if (user_choice == '2') and (computer_choice == 'Gun'):
        print("You win")
        player_score += 5
        if player_score > player_high_score:
            player_high_score = player_score


    if user_choice == '2' and computer_choice == 'Snake' :
        print("You Lost")
        computer_score += 5
        if computer_score > computer_high_score:
            computer_high_score = computer_score

    if user_choice == '2' and computer_choice == 'Water':
        print("Tie")
        player_score += 1
        computer_score += 1
        if player_score > player_high_score or computer_score > computer_high_score :
            player_high_score = player_score
            computer_high_score = computer_score
            
########### GUN #################################### 
    if (user_choice == '3') and (computer_choice == 'Snake'):
        print("You win")
        player_score += 5
        if player_score > player_high_score:
            player_high_score = player_score


    if user_choice == '3' and computer_choice == 'Water' :
        print("You Lost")
        computer_score += 5
        if computer_score > computer_high_score:
            computer_high_score = computer_score

    if user_choice == '3' and computer_choice == 'Gun':
        print("Tie")
        player_score += 1
        computer_score += 1
        if player_score > player_high_score or computer_score > computer_high_score :
            player_high_score = player_score
            computer_high_score = computer_score

    
    if user_choice != '1' and user_choice !='2' and user_choice != '3':
        print("Wrong Input")
        
            
    i += 1
print(f" Your Score: {player_high_score} \n Computer Score: {computer_high_score}")
if player_high_score > computer_high_score:
    print("Congratulations You Made it !")
elif player_high_score < computer_high_score:
    print("Oops Computer Won ")
else:
    print("Its a Tie !")
    
    
```
